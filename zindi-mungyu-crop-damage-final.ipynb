```python
import pandas as pd
from PIL import Image
import torch
from torchvision import models, transforms
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from torch.utils.data import Dataset, DataLoader

class CustomDataset(Dataset):
    def __init__(self, dataframe, transform=None):
        self.data_frame = dataframe  # 변경: csv_file 대신 dataframe 사용
        self.transform = transform

    def __len__(self):
        return len(self.data_frame)

    def __getitem__(self, idx):
        img_path = self.data_frame.iloc[idx, 3]
        image = Image.open(img_path)
        label = self.data_frame.iloc[idx, 1]

        if self.transform:
            image = self.transform(image)

        return image, label





##데이터 증강 (과적합막기용)
transform = transforms.Compose([
    transforms.Resize(256),
    transforms.RandomHorizontalFlip(),  # 수평 방향으로 무작위 뒤집기
    transforms.RandomRotation(10),      # 무작위로 회전
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])


###데이터 증강과 validatation 둘다 해야함 (early stopping)
###이대회에서는 데이터증강을할때 상하반전보다는 좌우반전이 더 났다 (이걸 확인하기위해서는 꼭 image 를 찍어봐야한
###Random rotation 도 도움될수있다 (식물이 다 서있는게 아니라 누워있는애들도 있기떄문이다)

train_df, valid_df = train_test_split(pd.read_csv('/kaggle/working/train123.csv'), test_size=0.2, random_state=42)

train_dataset = CustomDataset(dataframe=train_df, transform=transform)
valid_dataset = CustomDataset(dataframe=valid_df, transform=transform)

# DataLoader 설정 - 학습용과 검증용
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
valid_loader = DataLoader(valid_dataset, batch_size=32, shuffle=False)

# 사전 훈련된 모델 로드 및 수정
model = models.resnet18(pretrained=True)
num_ftrs = model.fc.in_features
model.fc = nn.Linear(num_ftrs, 5)  # '5'는 클래스 수에 따라 변경

# 손실 함수 및 옵티마이저 설정
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.001, momentum=0.9)

# GPU 사용 설정
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
model.to(device)

# Early Stopping 관련 설정
patience = 5
best_loss = float('inf')
patience_counter = 0

# 모델 훈련 및 Early Stopping
for epoch in range(10):
    model.train()
    running_loss = 0.0
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()

    # 검증 단계
    model.eval()
    valid_loss = 0.0
    with torch.no_grad():
        for inputs, labels in valid_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            valid_loss += loss.item()

    valid_loss /= len(valid_loader)
    print(f'Epoch {epoch + 1}, Training Loss: {running_loss / len(train_loader)}, Validation Loss: {valid_loss}')

    # Early Stopping 체크
    if valid_loss < best_loss:
        best_loss = valid_loss
        patience_counter = 0
    else:
        patience_counter += 1
        if patience_counter >= patience:
            print('Early stopping triggered')
            break

print('훈련 완료')

# 모델 저장
torch.save(model.state_dict(), '/kaggle/working/model.pth')

#학습횟수늘리기 => 도움이 많이되지만 early stopping 은 꼭걸어야한다 (e.g. epoch 늘리기)
```

```python
from PIL import Image
import torch
from torchvision import models, transforms
from torch.utils.data import Dataset, DataLoader

# Custom Dataset 클래스
class CustomDataset(Dataset):
    def __init__(self, csv_file, transform=None):
        self.data_frame = pd.read_csv(csv_file)
        self.transform = transform

    def __len__(self):
        return len(self.data_frame)

    def __getitem__(self, idx):
        img_path = self.data_frame.iloc[idx, 2]
        image = Image.open(img_path)

        if self.transform:
            image = self.transform(image)

        return image

# 데이터 변환 설정
transform = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

# Custom Dataset 인스턴스 생성
testset = CustomDataset(csv_file='/kaggle/working/test123.csv', transform=transform)

# DataLoader 설정
testloader = DataLoader(testset, batch_size=32, shuffle=False)

# 사전 훈련된 모델 로드
model = models.resnet18()
num_ftrs = model.fc.in_features
model.fc = nn.Linear(num_ftrs, 5)  # '5'는 클래스 수에 따라 변경
model.load_state_dict(torch.load('model.pth'))

# GPU 사용 설정
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
model.to(device)
model.eval()  # 평가 모드

import torch.nn.functional as F  # F를 임포트
from tqdm import tqdm
predictions = []
with torch.no_grad():
    for i, data in enumerate(tqdm(testloader, desc="Predicting")):
        data = data.to(device)
        outputs = model(data)
        probabilities = F.softmax(outputs, dim=1)  # softmax를 사용하여 확률 계산 (모델예측값을 logit 값으로 변환)
        predictions.extend(probabilities.cpu().numpy())

# 확률 값 출력 또는 저장
print(predictions)
```
