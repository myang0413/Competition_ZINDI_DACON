```python
import duckdb 
train_path = '/kaggle/input/daconctr/train.csv'

con = duckdb.connect()

train = con.query(f"""(SELECT *
                        FROM read_csv_auto('{train_path}')
                        WHERE Click = 0
                        ORDER BY random()
                        LIMIT 30000)
                        UNION ALL
                        (SELECT *
                        FROM read_csv_auto('{train_path}')
                        WHERE Click = 1
                        ORDER BY random()
                        LIMIT 30000)""").df()


con.close()
```

```python
train
```

```python
test=pd.read_csv("/kaggle/input/daconctr/test.csv")
test
```

```python
#train 랑 test 를 안묶고 처리하기
train2 = train.drop(columns = ["ID", "Click"])
test2 = test.drop(columns = ["ID"])
train2
```

```python
#데이터 대용량이라 concat 못했을때 label encoder 하는법 
from tqdm import tqdm
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for column in tqdm(train2.select_dtypes(include=['object']).columns):
    #column 하나마다 묶어버리는것 (한꺼번에 묶어버린것)
    combined_data = pd.concat([train2[column], test2[column]], axis=0).astype(str)
    le.fit(combined_data)
    train2[column] = le.transform(train2[column].astype(str))
    test2[column] = le.transform(test2[column].astype(str))
```

```python
from catboost import CatBoostClassifier
cbc=CatBoostClassifier(verbose=100,eval_metric="AUC")
cbc.fit(train2,train["Click"])
```

```python
result=cbc.predict_proba(test2)
result
```
